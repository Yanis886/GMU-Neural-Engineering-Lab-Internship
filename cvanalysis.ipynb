```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
readings: [(str,np)] = []
```

```python
for file in os.listdir("data"):
    readings.append([file, pd.read_excel(pd.ExcelFile("data/"+file)).to_numpy()])
```

```python
for index in range(len(readings)):
    a = np.copy(readings[index][1])
    col0 = a[:, 0]  # first column
    col1 = a[:, 1]  # second column
    col2 = a[:, 2]  # third column
    readings[index][1] = np.array([col0,col1,col2])

```

```python
colors = ["red", "green", "blue", "purple", "yellow", "black", "pink"]
for i in range(len(readings)):
    plt.plot(readings[i][1][0],readings[i][1][1], color = colors[i], label = readings[i][0][:-5] )
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Impedance (Ω)')
plt.legend()
plt.title("Impedance vs Frequency of Different Capacitors")
plt.savefig('graphs/zvfreq.png')
```

```python
colors = ["red", "green", "blue", "purple", "yellow", "black", "pink"]
plt.clf()
for i in range(len(readings)):
    plt.plot(readings[i][1][0],readings[i][1][2], color = colors[i], label = readings[i][0][:-5] )
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (Deg)')
plt.legend()
plt.title("Phase vs Frequency of Different Capacitors")
plt.savefig('graphs/phzvfreq.png')
```
